```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

data=df

# Prepare data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))

# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 60
X, y = create_sequences(scaled_data, seq_length)

# Split data
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(seq_length, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile and train
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, batch_size=32, epochs=10)

# Predict
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
predictions

# Evaluate
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"RMSE: {rmse}")
```

```python
from sklearn.ensemble import IsolationForest

# Prepare data
features = data[['Volume', 'Close', 'High', 'Low']]
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)

# Train Isolation Forest
model = IsolationForest(contamination=0.01)
model.fit(scaled_features)

# Detect anomalies
data['Anomaly'] = model.predict(scaled_features)
anomalies = data[data['Anomaly'] == -1]
print(anomalies)
```

```python
# Calculate daily returns
data['Daily Return'] = data['Close'].pct_change()

# Calculate rolling volatility
data['7D Volatility'] = data['Daily Return'].rolling(window=7).std() * np.sqrt(7)
data['30D Volatility'] = data['Daily Return'].rolling(window=30).std() * np.sqrt(30)

# Plot volatility
import matplotlib.pyplot as plt
plt.plot(data['Date'], data['7D Volatility'], label='7D Volatility')
plt.plot(data['Date'], data['30D Volatility'], label='30D Volatility')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Label data
data['Trade Label'] = np.where(data['Daily Return'] > 0, 1, 0)

# Prepare features and labels
features = data[['Volume', 'Daily Return', '7D Volatility']]
labels = data['Trade Label']

# Split data
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)

# Train Random Forest
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Evaluate
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
import pandas as pd
import requests
from textblob import TextBlob
import matplotlib.pyplot as plt

# Get news headlines
def get_news_headlines(api_key, query):
    url = f"https://newsapi.org/v2/everything?q={query}&apiKey={api_key}"
    response = requests.get(url)
    articles = response.json().get('articles', [])
    headlines = [article['title'] for article in articles]
    return headlines

api_key = "ad6a77d6dd8d4e7d86be6141c5da29b0"
enjin_headlines = get_news_headlines(api_key, "Enjin Coin")

# Perform sentiment analysis
def get_sentiment(text):
    if isinstance(text, str):
        return TextBlob(text).sentiment.polarity
    return 0

news_data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=len(enjin_headlines), freq='D'),
    'News Headline': enjin_headlines
})
news_data['Sentiment'] = news_data['News Headline'].apply(get_sentiment)

# Merge sentiment data with price data
data['Date'] = pd.to_datetime(data['Date'])
news_data['Date'] = pd.to_datetime(data['Date'])
merged_data = pd.merge(data, news_data, on='Date', how='left')
merged_data['Sentiment'] = merged_data['Sentiment'].fillna(0)

# Calculate correlation
merged_data['Daily Return'] = merged_data['Close'].pct_change()
correlation = merged_data['Sentiment'].corr(merged_data['Daily Return'])
print(f"Correlation between sentiment and daily returns: {correlation}")

# Visualize the results
plt.figure(figsize=(10, 6))
plt.plot(merged_data['Date'], merged_data['Sentiment'], label='Sentiment', color='blue')
plt.plot(merged_data['Date'], merged_data['Daily Return'], label='Daily Return', color='orange')
plt.legend()
plt.title("Sentiment vs. Daily Returns for Enjin Coin (ENJ)")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```
